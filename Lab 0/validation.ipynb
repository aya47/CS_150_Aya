```python
# Input: two sequences, strings
  # DNA like: should consist of A, T, C, G and indel symbol -

# Commands
  # u = update sequences (re-enter)
  # s = score (score the alignment)
  # q = quit (stop the process)

# Scoring
  # Report the number of matches in two sequences
    # if one string is shorter than the other -> matching stops after shorter sequence is done
    # if an indel exists -> no match
```

```python
# is the input a valid DNA sequence?

# Input
dna = input("Please enter DNA sequence (using uppercase letters A, T, C, G and an indel symbol): ")

if len(dna) == 0:
  print(dna, "is not a valid DNA sequence")

# Initialize counter
counter = 0
# Validate
for ch in dna:
  if ch in "ATCG-":
    counter += 1
    continue
  else:
    print(dna, "is not a valid DNA sequence")
    break

# Confirm validation
if counter == len(dna) and counter != 0:
  print(dna, "is a valid DNA sequence")
```
