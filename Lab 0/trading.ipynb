```python
# Prices have only one digit
# Stay the same the whole day
# History = one long string of numbers
# if prices is the string, prices[i] is the price of the item on day i

# Input: Price history
prices = input("Enter price history: ")

# Analyze history
n = len(prices)
max = 0
for i in range(n):
  print(i)
  for j in range(i+1, n):
    profit = int(prices[j]) - int(prices[i])
    if profit > max:
      max = profit

# Output profit
print("Profit: ", max)
```
