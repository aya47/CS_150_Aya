```python
# objective: calculate score of alignment

# Input:
dna_1 = input("Please enter DNA sequence (using uppercase letters A, T, C, G and an indel symbol): ")
dna_2 = input("Please enter DNA sequence (using uppercase letters A, T, C, G and an indel symbol): ")

# n is the length i am going for

if len(dna_1) <= len(dna_2):
    n = len(dna_1)
else:
    n = len(dna_2)

# Initialize score
score = 0
for i in range(n):
  if dna_1[i] == dna_2[i]:
    score += 1
  else:
    continue

# Output
if score == 1:
  print(score, " match found between ", dna_1, " and ", dna_2)
else:
  print(score, " matches found between ", dna_1, " and ", dna_2)
```
