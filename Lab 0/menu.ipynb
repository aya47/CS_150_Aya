```python
# Input: two sequences, strings
  # DNA like: should consist of A, T, C, G and indel symbol -

# Commands
  # u = update sequences (re-enter)
  # s = score (score the alignment)
  # q = quit (stop the process)

# Scoring
  # Report the number of matches in two sequences
    # if one string is shorter than the other -> matching stops after shorter sequence is done
    # if an indel exists -> no match
```

```python
def menu():
  print("Select one of the following commands: ")
  print("'u' to update sequences")
  print("'s' to score the alignment")
  print("'q' to quit")

print("Hello and welcome to DNA sequence alignment program!")

while True:
  menu()
  str = input()
  if str == 'q':
    print("Thank you! Goodbye! ")
    break
```
