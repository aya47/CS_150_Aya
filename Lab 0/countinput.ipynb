```python
# Input: line of text
text = input()

# Initialize counter
count = 0

# Compute number of characters over a loop
for ch in text:
  # Exclude spaces, periods, exclamation or commas
  if ch in " .!,":
    continue
  count += 1

# Output: number of characters, excluding spaces, periods, exclamation or commas
print(count)

# Test:
  # Input: Listen, Mr. Jones, calm down.
  # Output: 21
```
