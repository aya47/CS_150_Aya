```python
# Input
t = int(input("Enter the average daily temperature: "))

# Initialize variables to store values in
cooling_list = []
heating_list = []
# Initial Condition
if t < 60 and t > -459:
  heating_list.append(t)
  #print(heating_list)
elif t > 80:
  cooling_list.append(t)
  #print(cooling_list)

# Loop to keep collecting values until t < -459
while t > -459:
  t = int(input("Enter the average daily temperature: "))
  # condition: if t_ave < 60 -> heating day
            #if t_ave > 80 -> cooling day
  #print(t)
  if t < 60 and t > -459:
    heating_list.append(t)
    #print(heating_list)
  elif t > 80:
    cooling_list.append(t)
    #print(cooling_list)

# Output
print("Heating days: ", len(heating_list))
print("Cooling days: ", len(cooling_list))
```
