```python
# Functions
## MENU
def menu():
  print("Select one of the following commands: ")
  print("'u' to update sequences")
  print("'s' to score the alignment")
  print("'q' to quit")

## DNA INPUT
def input_dna():
  global dna_1
  global dna_2
  dna_1 = input("Please enter DNA sequence (using uppercase letters A, T, C, G and an indel symbol): ")
  dna_2 = input("Please enter DNA sequence (using uppercase letters A, T, C, G and an indel symbol): ")

  print("You entered: ")
  print(dna_1)
  print(dna_2)

## ALIGNMENT SCORE CALCULATION
def score():
  if len(dna_1) <= len(dna_2):
    n = len(dna_1)
  else:
    n = len(dna_2)

  # Initialize score
  score = 0
  for i in range(n):
    if dna_1[i] == dna_2[i]:
      score += 1
    else:
      continue
  # Output
  if score == 1:
    print(score, " match found between ", dna_1, " and ", dna_2)
  else:
    print(score, " matches found between ", dna_1, " and ", dna_2)

## DNA INPUT VALIDATION
def validate(dna):
  flag = 0
  if len(dna) == 0:
    flag = 1
  # Validate
  for ch in dna:
    if ch in "ATCG-":
      continue
    else:
      flag = 1
      break
  return flag
```

```python
print("Hello and welcome to DNA sequence alignment program!")
input_dna()

while True:
  menu()
  str = input()
  if str == 'q':
    print("Thank you! Goodbye! ")
    break
  elif str == 'u':
    input_dna()
  elif str == 's':
    if validate(dna_1) == 0 and validate(dna_2) == 0:
      score()
    else:
      print("Invalid DNA sequences entered, please re-enter sequences")
  else:
    continue
```
